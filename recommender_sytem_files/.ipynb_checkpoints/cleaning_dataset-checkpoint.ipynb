```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import surprise
```

```python
users = pd.read_csv('BX-Users.csv',sep=';') 
items = pd.read_csv('BX-Books.csv',sep=';',error_bad_lines=False)
ratings = pd.read_csv('BX-Book-Ratings.csv',sep=';')
ratings = ratings[ratings['Book-Rating'] > 0]
```

```python
ratings = ratings[ratings['ISBN'].isin(items['ISBN'])]

density = (float(len(ratings))/(len(np.unique(ratings['User-ID']))*len(np.unique(ratings['ISBN']))))*100
print(density) 
print ( len(ratings['User-ID'].unique()) , len(ratings['ISBN'].unique()) )
```
